```python
import pandas as pd
import csv

file_path = 'H://file csv//th5.csv'

# Mở file với mã hóa 'utf-8' và bỏ qua các lỗi mã hóa
with open(file_path, mode='r', encoding='utf-8', errors='replace') as file:
    content = file.read()

# Lưu nội dung vào một file tạm thời
temp_path = 'temp_file.csv'
with open(temp_path, mode='w', encoding='utf-8') as temp_file:
    temp_file.write(content)

# Đọc file tạm thời vào dataframe
df = pd.read_csv(temp_path)


# print ten cac cot
print(df.columns)
```
