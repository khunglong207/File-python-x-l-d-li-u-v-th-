```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ng dn n cc tp CSV
file_paths = [
    'H://file csv//th41.csv',
    'H://file csv//th42.csv',
    'H://file csv//th43.csv',
    'H://file csv//th44.csv'
]

# c d liu t cc tp CSV v lm sch d liu
data_frames = []
for file in file_paths:
    df = pd.read_csv(file)
    df = df.apply(pd.to_numeric, errors='coerce')  # Chuyn i cc gi tr thnh s, cc gi tr khng phi s s l NaN
    df = df.dropna()  # Loi b cc hng c gi tr NaN
    data_frames.append(df)

# To  th 3D
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

# V tng tp d liu
colors = ['r', 'g', 'b', 'y']  # Mu sc cho cc tp d liu khc nhau
for i, df in enumerate(data_frames):
    df = df.iloc[:50000]  # Ch v n dng th 50000
    df = df.iloc[::30]  # Ly mu 1/50 s im d liu  gim  dy
    x = df.iloc[:, 0]
    y = df.iloc[:, 1]
    z = df.iloc[:, 2]
    ax.scatter(x, y, z, label=f'Cf_{i+1}', color=colors[i % len(colors)], s=12, edgecolor='k', alpha=0.6)

# t nhn cho cc trc
ax.set_xlabel('X [meters]', labelpad=8, fontsize=8)
ax.set_ylabel('Y [meters]', labelpad=8, fontsize=8)
ax.set_zlabel('Z [meters]', labelpad=2, fontsize=8)

# t tn cho  th
ax.set_title('Do thi 3D mo phong cua TH4', fontsize=15, pad=15)

# Hin th  th
ax.legend(fontsize=10)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Thit lp qu o bay lemniscate
def lemniscate_trajectory(t, trajectory_start_position):
    a = 0.5
    c = 0.3
    omega = 0.75*np.tanh(0.1*t)
    pxr = trajectory_start_position[0] + a*np.cos(omega*t) - a
    pyr = trajectory_start_position[1] + a*np.sin(omega*t)
    pzr = trajectory_start_position[2] + c*np.sin(omega*t)
    return pxr, pyr, pzr

# To d liu qu o cho 4 v tr bt u khc nhau
start_positions = [
    [0, 0, 0],
    [1, 0, 0],
    [0, 1, 0],
    [1, 1, 0]
]

# To  th 3D
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# V tng qu o lemniscate
colors = ['r', 'g', 'b', 'y']
t = np.linspace(0, 100, 5000)  # Thay i phm vi v s lng im nu cn

for i, start_pos in enumerate(start_positions):
    x_data, y_data, z_data = lemniscate_trajectory(t, start_pos)
    ax.plot(x_data, y_data, z_data, label=f'Cf_{i+1}', color=colors[i], linewidth=2)

# t nhn cho cc trc
ax.set_xlabel('X [meters]', labelpad=8, fontsize=8)
ax.set_ylabel('Y [meters]', labelpad=8, fontsize=8)
ax.set_zlabel('Z [meters]', labelpad=2, fontsize=8)

# t tn cho  th
ax.set_title('Do thi 3D theo ly thuyet cua TH4', fontsize=15, pad=15)

# Hin th  th
ax.legend(fontsize=10)
plt.show()
```
